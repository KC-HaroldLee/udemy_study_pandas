#####  pairplot
###### 데이터 프레임에서 변수간 관계를 간단하게 확인하는 방법으로 scatter plot을 그려 시각적으로 확인하는 방법이 있다. 변수가 여러 개일 때에는 for 문으로 변수들을 조합하여 각 조합별 scatter plot을 그릴 수도 있지만, seaborn 패키지를 이용하면 간단하게 그림을 그릴 수 있다. https://m.blog.naver.com/saftyzon/222387168800

```python
sns.pairplot(cancer_df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```
