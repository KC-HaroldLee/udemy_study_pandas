#### 아래는 그냥 정리 한 것

```python
x = 'Date'

y_list = ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price']
y_idx = 0
y = y_list[y_idx]

label = 'Bitcoin Price'

linewidth = 3

figsize = (14, 6)

investments_df.plot(x=x, y=y, label=label, linewidth=linewidth, figsize=figsize)

ylabel = 'Price [$]'
title = 'bit pasdasd'
legend_loca_list = ["upper left", "upper center", "upper right", "lower left", "lower center", "lower right"]
legend_loca_idx = 1
legend_loca = legend_loca_list[legend_loca_idx]

plt.ylabel(ylabel)
plt.title(title)
plt.legend(loc=legend_loca)
plt.grid()
```

```python
import yfinance as yf
```

```python
# 타입 잘 기억하기
investments_list = ["BTC-USD"]

start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2023, 12, 31)

df = yf.download(investments_list, start = start, end = end)
df
```

```python
df.info()
```

#### 인덱스를 바꾼다(기본으로) x를 빼기위함같긴함

```python
df.reset_index(inplace = True)
df
```

##### 심심하니까 한 번 갖고 놀기

```python
df.plot(x = 'Date', y = ['High', 'Low'], linewidth = 2, figsize = (14,6))
plt.ylabel('Price')
plt.title('Crypto Prices')
plt.legend(['HIGH-V', 'LOW-V'], loc='upper left')
plt.grid()
```

#### subplot # 따로 지정하기 어려운 부분들이 생김

```python
df.plot(x = 'Date', y = ['High', 'Low'], title='subplot test', subplots = True, grid = True, linewidth = 2, figsize = (14,6))
# plt.ylabel('Price')
# plt.title('Crypto Prices')
# plt.legend(['HIGH-V', 'LOW-V'], loc='upper left')
# plt.grid()
```

```python
df.plot(x = 'Date', y = ['High', 'Low'], 
        title='subplot test', 
        subplots = True, 
        grid = True, 
        linewidth = 2, 
        figsize = (14,6),
        )
```

#### 아래는 예제인데 추가 공부가 필요해 보여서 이것저것 적어 놓음

```python
mu = daily_return_df['BTC'].mean() # mean of distribution
sigma = daily_return_df['BTC'].std() # standard deviation of distribution

print("mu : ", mu)
print("sigma : ", sigma)
```
