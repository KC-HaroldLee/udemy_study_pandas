```python
import pandas as pd
```

#### 간단한 Concat이다. 그러니까 셀 하나에 그냥 넣을거임...

```python
raw_data_1 = {'Bank Client ID': ['1', '2', '3', '4', '5'],
            'First Name': ['Nancy', 'Alex', 'Shep', 'Max', 'Allen'],
            'Last Name': ['Rob', 'Ali', 'George', 'Mitch', 'Steve']}

bank_1_df = pd.DataFrame(raw_data_1, columns = ['Bank Client ID', 'First Name', 'Last Name'])

raw_data_2 = {
        'Bank Client ID': ['6', '7', '8', '9', '10'],
        'First Name': ['Bill', 'Dina', 'Sarah', 'Heather', 'Holly'],
        'Last Name': ['Christian', 'Mo', 'Steve', 'Bob', 'Michelle']}
bank_2_df = pd.DataFrame(raw_data_2)

bank_all_df = pd.concat([bank_1_df, bank_2_df])
bank_all_df
```

#### 인덱스를 없앤다. 이건 첨알았네

```python
bank_all_df = pd.concat([bank_1_df, bank_2_df], ignore_index = True)
bank_all_df
```

###### 왜 수강생 리뷰가 좋은거지?

```python
# bank_all_df = bank_2_df.append(bank_1_df, ignore_index = True) ### <<< 없어진 함수
bank_all_df = bank_2_df._append(bank_1_df, ignore_index = True)
bank_all_df
```

##### 특이한 기능이다만 보기에만 좋고 실제 데이터를 다루는 데에 어떤지는 봐야한다.

```python
bank_all_df = pd.concat([bank_1_df, bank_2_df], keys = ["Customers Group 1", "Customers Group 2"])
bank_all_df
```

##### 바로 확인해보자
##### 일단 iterrow()을 할경우 row_idx(내표현)이 두개의 요소를 가진 튜플이 된다. 그리고 그 값은 name에서도 찾을 수 있다. 나름 그룹 서칭할 때 편할듯? 한데 그냥 조건서칭이 빠를거 같기도. 혹은 그냥 DB의 테이블처럼 컬럼 하나 추가 하는 것이 더 많은 Select를 할 수 있을 거 같으니 이런것도 있다는 것만 기억해두자. 

```python
for row_idx, row_data in bank_all_df.iterrows() :
    print(row_idx)
    if int(row_idx[1]) % 5 == 0 :
        print(row_data)
    break
```
